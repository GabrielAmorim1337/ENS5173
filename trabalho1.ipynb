```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Suponha que você tenha os dados dos poluentes em um DataFrame chamado "df_poluentes"

# Lista de poluentes
poluentes = ['PM', 'SO2', 'NOX', 'CO','TOC']

# Para cada poluente, criando gráficos
for poluente in poluentes:
    # Criando os valores aleatórios e aleatórios sazonais
    # aleátorio sazonal é a média entre os valores variáveis de atividade com valores aleatórios que variam entre a atividade máxima e mínima multiplicado pela sazonalidade 
    
    df['aleatorio'] = (atvMax - atvMin) * np.random.rand(n) + atvMin
    aleatorioSazonal = (df['atividade'] +  df['sazonalidade']*df['atividade'].mean())
    df['aleatorioSazonal'] = aleatorioSazonal
    
    # Criando os subplots
    fig, ax = plt.subplots(2)
    
    # Plotando o gráfico dos valores aleatórios com sazonalidade
    ax[1].plot(df['Dia'], df['aleatorioSazonal'], color='midnightblue')
    ax[1].set_title('Valores Aleatórios Sazonais para {}'.format(poluente))
    ax[1].set_xlabel('Dia')
    ax[1].set_ylabel('Valor Aleatório Sazonal')
    ax[1].grid(True)
    
    # Plotando o gráfico da sazonalidade
    ax[0].plot(df['sazonalidade'] * df['atividade'].mean(), color='midnightblue')
    ax[0].set_title('Sazonalidade para {}'.format(poluente))
    ax[0].set_xlabel('Hora')
    ax[0].set_ylabel('Sazonalidade')
    ax[0].grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python
# Lista de poluentes
poluentes = ['PM', 'SO2', 'NOX', 'CO', 'TOC']

# Criar um DataFrame vazio para armazenar os resultados para Dry Process Kiln
resultados_dry_process_kiln = pd.DataFrame(index=df.index)

# Multiplicar os valores de cada poluente pela atividade para Dry Process Kiln
for poluente in poluentes:
    resultados_dry_process_kiln[poluente] = df['emisaleatoria'] * emis.loc[0, poluente]

# Visualizar a tabela para Dry Process Kiln
print("Tabela para o tipo de fonte: Dry Process Kiln")
resultados_dry_process_kiln
```

```python
# Criar um DataFrame vazio para armazenar os resultados para Prehearter / PreCalciner
resultados_preheater_precalciner = pd.DataFrame(index=df.index)

# Multiplicar os valores de cada poluente pela atividade para Prehearter / PreCalciner
for poluente in poluentes:
    resultados_preheater_precalciner[poluente] = df['emisaleatoria'] * emis.loc[1, poluente]

# Visualizar a tabela para Prehearter / PreCalciner
print("Tabela para o tipo de fonte: Prehearter / PreCalciner")
resultados_preheater_precalciner
```

```python
# Criar um DataFrame vazio para armazenar os resultados para Clincker Cooker
resultados_clincker_cooker = pd.DataFrame(index=df.index)

# Multiplicar os valores de cada poluente pela atividade para Clincker Cooker
for poluente in poluentes:
    resultados_clincker_cooker[poluente] = df['emisaleatoria'] * emis.loc[2, poluente]

# Visualizar a tabela para Clincker Cooker
print("Tabela para o tipo de fonte: Clincker Cooker")
resultados_clincker_cooker
```

```python
# Criar um DataFrame vazio para armazenar os resultados para Clincker Cooler
resultados_clincker_cooler = pd.DataFrame(index=df.index)

# Multiplicar os valores de cada poluente pela atividade para Clincker Cooler
for poluente in poluentes:
    resultados_clincker_cooler[poluente] = df['emisaleatoria'] * emis.loc[3, poluente]

# Visualizar a tabela para Clincker Cooler
print("Tabela para o tipo de fonte: Clincker Cooler")
resultados_clincker_cooler
```

```python
import matplotlib.pyplot as plt

# Definir cores para cada processo
cores = ['darkcyan', 'crimson', 'springgreen', 'y']

# Criar função para plotar gráfico para cada poluente em cada processo
def plotar_graficos_por_poluente(resultados, tipo_fonte, cor):
    plt.figure(figsize=(15, 10))
    # Iterar sobre cada poluente
    for i, poluente in enumerate(poluentes, start=1):
        if not resultados.empty:  # Verificar se há dados antes de plotar
            plt.subplot(3, 2, i)
            plt.plot(resultados.index, resultados[poluente], color=cor)
            plt.title(f'Emissões de {poluente} - {tipo_fonte}')
            plt.xlabel('Tempo (horas)')
            plt.ylabel('Emissões')
            plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plotar gráficos individuais para cada poluente em cada processo
plotar_graficos_por_poluente(resultados_dry_process_kiln, 'Dry Process Kiln', cores[0])
plotar_graficos_por_poluente(resultados_preheater_precalciner, 'Prehearter / PreCalciner', cores[1])
plotar_graficos_por_poluente(resultados_clincker_cooker, 'Clincker Cooker', cores[2])
plotar_graficos_por_poluente(resultados_clincker_cooler, 'Clincker Cooler', cores[3])

#gráficos das emissões de cada poluente em cada processo em kg/hora
```

```python
# Somar os poluentes emitidos em cada processo
# Convertendo as unidades encontradas
# Resultados estão em kg/ano, pois é a soma dos resultados das 8736 horas do ano
# Convertendo os dados de kg/ano para tonelas/ano para melhor visualizar os resultados



soma_por_processo = {
    'Dry Process Kiln': resultados_dry_process_kiln.sum() / 1000,
    'Prehearter / PreCalciner': resultados_preheater_precalciner.sum() / 1000,
    'Clincker Cooker': resultados_clincker_cooker.sum() / 1000,
    'Clincker Cooler': resultados_clincker_cooler.sum() / 1000
}

# Mostrar a soma dos poluentes em cada processo
print("Soma dos poluentes produzidos em cada processo (em toneladas por ano):")
for processo, soma in soma_por_processo.items():
    print(f"{processo}:")
    print(soma)
    print()

# Somar todos os poluentes de todos os processos e converter para toneladas
soma_total = (resultados_dry_process_kiln.sum() + resultados_preheater_precalciner.sum() + resultados_clincker_cooker.sum() + resultados_clincker_cooler.sum()) / 1000

# Mostrar a soma total dos poluentes
print("Soma total dos poluentes produzidos de todos os processos (em toneladas por ano):")
print(soma_total)
```
