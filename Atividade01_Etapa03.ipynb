#### Histórico de 2008 a 2017 da capacidade elétrica instalada das diferentes fontes de geração de energia elétrica no Brasil.
###### Fonte : Dachery J. ,2019)


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Função para receber dados de produção elétrica e datas

data = np.arange(2008,2017)
producao = np.array([104, 113, 117, 120, 126, 133, 140, 150, 157])
```

```python
# Calculando média e desvio padrão
media = np.mean(producao)
desvio_padrao = np.std(producao)
```

```python
# Gerando números aleatórios para os anos de 2017 a 2025
data_estendida = np.arange(2017, 2025)
producao_estendida = np.random.normal(media, desvio_padrao, len(data_estendida))
```

```python
#Exibindo os números aleatórios gerados
print("Capacida de produção de energia gerada aleatoriamente até 2025:")
for ano, valor in zip(data_estendida, producao_estendida):
    print(f"Ano de {ano}: {valor:.2f} megawatts de capacidade de produção elétrica")
```

```python
# Plotagem do gráfico
plt.figure(figsize=(10, 6))
plt.plot(data, producao, marker='o', linestyle='-', label='Dados Obtidos  ')
plt.plot(data_estendida, producao_estendida, marker='o', linestyle='-', label='Dados Modelados ')
plt.title('Capacidade Instalada de produção de elétrica (Megawatts)')
plt.xlabel('Ano')
plt.ylabel('Capacidade de Produção (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# Exibição do gráfico
plt.show()
```
