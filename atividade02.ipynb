```python
from datetime import datetime, timedelta
from siphon.simplewebservice.wyoming import WyomingUpperAir
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Função para obter dados de temperatura em uma data e estação específicas
def get_temperature_data(date, station):
    df = WyomingUpperAir.request_data(date, station)
    return df

# Função para obter as temperaturas em 0h e 12h de um determinado dia
def get_day_temperatures(date, station):
    temperatures = []
    for hour in [0, 12]:
        date_with_hour = date.replace(hour=hour)
        df = get_temperature_data(date_with_hour, station)
        # Filtrando para altitude mais próxima ao solo (5 m de altitude)
        df_sea_level = df[df['height'] == 5]
        if not df_sea_level.empty:
            # Extraindo a temperatura (em Celsius) ao nível do mar
            temperature = df_sea_level['temperature'].values[0]
            temperatures.append((date_with_hour, temperature))
        else:
            temperatures.append((date_with_hour, None))  # Sem dados disponíveis
    return temperatures

# Função para obter temperaturas em 0h e 12h para um número de dias consecutivos
def get_consecutive_days_temperatures(start_date, station, num_days):
    all_temperatures = []
    for i in range(num_days):
        date = start_date + timedelta(days=i)
        day_temperatures = get_day_temperatures(date, station)
        all_temperatures.extend(day_temperatures)
    return all_temperatures

# Definindo a estação e as datas de início para verão e inverno
station = 'SBFL'
summer_start_date = datetime(2023, 1, 10)  # Inverno (Janeiro)
winter_start_date = datetime(2023, 7, 10)  # Verão (Julho)

# Obtendo temperaturas para 5 dias consecutivos no verão e no inverno
summer_temperatures = get_consecutive_days_temperatures(summer_start_date, station, 5)
winter_temperatures = get_consecutive_days_temperatures(winter_start_date, station, 5)

# Imprimindo os resultados
print("Temperaturas no verão (5 dias consecutivos):")
for date, temp in summer_temperatures:
    if temp is not None:
        print(f"{date.strftime('%Y-%m-%d %H:%M:%S')}: {temp:.2f} °C")
    else:
        print(f"{date.strftime('%Y-%m-%d %H:%M:%S')}: Sem dados disponíveis")

print("\nTemperaturas no inverno (5 dias consecutivos):")
for date, temp in winter_temperatures:
    if temp is not None:
        print(f"{date.strftime('%Y-%m-%d %H:%M:%S')}: {temp:.2f} °C")
    else:
        print(f"{date.strftime('%Y-%m-%d %H:%M:%S')}: Sem dados disponíveis")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from siphon.simplewebservice.wyoming import WyomingUpperAir

# Função para identificar os pontos de inversão de temperatura
def identify_inversions(df):
    temp = df['temperature'].values
    alt = df['height'].values
    pontos_inversao = np.where(np.diff(np.sign(np.diff(temp))) != 0)[0] + 1
    
    inversions = [(alt[i], temp[i]) for i in pontos_inversao if 
                  (temp[i] > temp[i - 1] and temp[i] > temp[i + 1]) or 
                  (temp[i] < temp[i - 1] and temp[i] < temp[i + 1])]
    
    return inversions

# Função para encontrar a camada limite
def find_boundary_layer(df):
    inversions = identify_inversions(df)
    if inversions:
        return inversions[0]  # A primeira inversão encontrada é considerada a camada limite
    return None

# Função para obter dados de temperatura
def get_temperature_data(date, station):
    df = WyomingUpperAir.request_data(date, station)
    return df

# Função para plotar os dados de temperatura com inversões e camada limite destacadas
def plot_temperature_with_boundary_layer(df, title, hour):
    plt.figure(figsize=(10, 6))
    plt.plot(df['temperature'], df['height'], label='Perfil de Temperatura', color='red')
    
    # Identificar inversões e camada limite
    inversions = identify_inversions(df)
    boundary_layer = find_boundary_layer(df)
    
    label_added = False
    for inversion in inversions:
        if not label_added:
            plt.scatter(inversion[1], inversion[0], color='black', label='Inversão de Temperatura', zorder=5)
            label_added = True
        else:
            plt.scatter(inversion[1], inversion[0], color='black', zorder=5)
    
    if boundary_layer:
        plt.axhline(boundary_layer[0], color='green', linestyle='--', label='Camada Limite', zorder=4)
    
    plt.gca().invert_yaxis()
    plt.xlabel('Temperatura (°C)')
    plt.ylabel('Altura (m)')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Função para obter temperaturas de 5 dias consecutivos
def get_consecutive_days_temperatures(start_date, station, num_days):
    temperatures = []
    for i in range(num_days):
        for hour in [0, 12]:
            date_with_hour = start_date + timedelta(days=i, hours=hour)
            df = get_temperature_data(date_with_hour, station)
            temperatures.append((date_with_hour, df))
    return temperatures

# Definindo a estação e as datas de início para verão e inverno
station = 'SBFL'
summer_start_date = datetime(2023, 1, 10)  # Verão (Janeiro)
winter_start_date = datetime(2023, 7, 10)  # Inverno (Julho)

# Obtendo temperaturas para 5 dias consecutivos no verão e no inverno
summer_temperatures = get_consecutive_days_temperatures(summer_start_date, station, 5)
winter_temperatures = get_consecutive_days_temperatures(winter_start_date, station, 5)

# Função para calcular a altura média da camada limite
def calculate_average_boundary_layer_height(temperatures):
    boundary_heights = []
    for date, df in temperatures:
        boundary_layer = find_boundary_layer(df)
        if boundary_layer:
            boundary_heights.append(boundary_layer[0])
    if boundary_heights:
        return np.mean(boundary_heights)
    return None

# Calculando a altura média da camada limite para verão e inverno
average_boundary_height_summer = calculate_average_boundary_layer_height(summer_temperatures)
average_boundary_height_winter = calculate_average_boundary_layer_height(winter_temperatures)

print(f"Altura média da camada limite no verão: {average_boundary_height_summer:.2f} m")
print(f"Altura média da camada limite no inverno: {average_boundary_height_winter:.2f} m")

# Plotando os dados de verão com inversões e camada limite destacadas
for date, df in summer_temperatures:
    hour = 0 if date.hour == 0 else 12
    plot_temperature_with_boundary_layer(df, f"Perfil de Temperatura - Verão {date.strftime('%Y-%m-%d')} ({hour}h)", hour)

# Plotando os dados de inverno com inversões e camada limite destacadas
for date, df in winter_temperatures:
    hour = 0 if date.hour == 0 else 12
    plot_temperature_with_boundary_layer(df, f"Perfil de Temperatura - Inverno {date.strftime('%Y-%m-%d')} ({hour}h)", hour)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Caminho para o arquivo CSV
meteo_file = 'C:\\Users\\HP\\Documents\\Gabriel.py\\ENS5173\\dados brutos\\METEO\\meteo.csv'

# Carregando os dados
meteo = pd.read_csv(meteo_file)

# Convertendo data e hora para datetime
meteo['datetime'] = pd.to_datetime(meteo[['year', 'month', 'day', 'hour']])

# Filtrando dados de 2023
meteo_2023 = meteo[meteo['year'] == 2023]

# Função para identificar o quadrante do vento
def wind_quadrant(direction):
    if 0 <= direction < 90:
        return '1º Quadrante (0° a 90°)'
    elif 90 <= direction < 180:
        return '2º Quadrante (90° a 180°)'
    elif 180 <= direction < 270:
        return '3º Quadrante (180° a 270°)'
    elif 270 <= direction < 360:
        return '4º Quadrante (270° a 360°)'
    else:
        return None

# Aplicando a função aos dados
meteo_2023['Wind Quadrant'] = meteo_2023['Wind Direction (degrees)'].apply(wind_quadrant)

# Filtrando valores indefinidos
meteo_2023 = meteo_2023[meteo_2023['Wind Quadrant'].notna()]

# Resumo estatístico dos parâmetros de superfície
albedo_summary = meteo_2023['albedo (month and wind direction dependent; 1.0 for hours before sunrise or after sunset) (non-dimensional)'].describe()
bowen_summary = meteo_2023['Bowen ratio (month and wind direction dependent) (non-dimensional)'].describe()
roughness_summary = meteo_2023['surface roughness length (month and wind direction dependent) (m)'].describe()

print("Resumo do Albedo em 2023:")
print(albedo_summary)
print("\nResumo da Razão de Bowen em 2023:")
print(bowen_summary)
print("\nResumo da Rugosidade em 2023:")
print(roughness_summary)

# Filtrando apenas quadrantes presentes nos dados
quadrants = meteo_2023['Wind Quadrant'].unique()
quadrant_labels = sorted(quadrants)

# Definindo cores
colors = {
    'albedo': 'seagreen',
    'bowen': 'red',
    'roughness': 'skyblue'
}

# Plotando os parâmetros de superfície por quadrante do vento
fig, axes = plt.subplots(3, 1, figsize=(12, 18), sharex=True)
fig.suptitle('Parâmetros de Superfície por Quadrante do Vento')

# Plotando Albedo
albedo_data = [meteo_2023[meteo_2023['Wind Quadrant'] == q]['albedo (month and wind direction dependent; 1.0 for hours before sunrise or after sunset) (non-dimensional)'] for q in quadrant_labels]
axes[0].boxplot(albedo_data, patch_artist=True, boxprops=dict(facecolor=colors['albedo']))
axes[0].set_xticks(range(1, len(quadrant_labels) + 1))
axes[0].set_xticklabels(quadrant_labels)
axes[0].set_ylabel('Albedo')
axes[0].set_title('Albedo por Quadrante do Vento')

# Plotando Razão de Bowen
bowen_data = [meteo_2023[meteo_2023['Wind Quadrant'] == q]['Bowen ratio (month and wind direction dependent) (non-dimensional)'] for q in quadrant_labels]
axes[1].boxplot(bowen_data, patch_artist=True, boxprops=dict(facecolor=colors['bowen']))
axes[1].set_xticks(range(1, len(quadrant_labels) + 1))
axes[1].set_xticklabels(quadrant_labels)
axes[1].set_ylabel('Razão de Bowen')
axes[1].set_title('Razão de Bowen por Quadrante do Vento')

# Plotando Rugosidade
roughness_data = [meteo_2023[meteo_2023['Wind Quadrant'] == q]['surface roughness length (month and wind direction dependent) (m)'] for q in quadrant_labels]
axes[2].boxplot(roughness_data, patch_artist=True, boxprops=dict(facecolor=colors['roughness']))
axes[2].set_xticks(range(1, len(quadrant_labels) + 1))
axes[2].set_xticklabels(quadrant_labels)
axes[2].set_ylabel('Rugosidade (m)')
axes[2].set_title('Rugosidade por Quadrante do Vento')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Analisando agora em octantes ao invés de quadrantes para obter uma melhor análise das condições atmosféricas no local

# Carregar os dados meteorológicos do arquivo CSV
meteo_file = 'C:\\Users\\HP\\Documents\\Gabriel.py\\ENS5173\\dados brutos\\METEO\\meteo.csv'
meteo = pd.read_csv(meteo_file)

# Remover valores de direção do vento iguais a 999
meteo = meteo[meteo['Wind Direction (degrees)'] != 999]

# Definir os limites dos octantes de vento em graus
octant_limits = [(0, 45), (45, 90), (90, 135), (135, 180), (180, 225), (225, 270), (270, 315), (315, 360)]

# Inicializar listas para armazenar as médias dos parâmetros e o número de amostras para cada octante
albedo_means = []
bowen_ratio_means = []
roughness_means = []
num_samples = []

# Iterar sobre os octantes de vento
for lower_limit, upper_limit in octant_limits:
    # Selecionar os dados dentro do intervalo de direção do vento atual
    octant_data = meteo[(meteo['Wind Direction (degrees)'] >= lower_limit) & (meteo['Wind Direction (degrees)'] < upper_limit)]
    
    # Calcular o número de amostras para este octante
    num_samples.append(len(octant_data))
    
    # Calcular as médias dos parâmetros para este octante
    albedo_mean = octant_data['albedo (month and wind direction dependent; 1.0 for hours before sunrise or after sunset) (non-dimensional)'].mean()
    bowen_ratio_mean = octant_data['Bowen ratio (month and wind direction dependent) (non-dimensional)'].mean()
    roughness_mean = octant_data['surface roughness length (month and wind direction dependent) (m)'].mean()
    
    # Adicionar as médias às listas
    albedo_means.append(albedo_mean)
    bowen_ratio_means.append(bowen_ratio_mean)
    roughness_means.append(roughness_mean)

# Definir os rótulos dos octantes
octant_labels = ['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']

# Definir os ângulos para cada octante da rosa dos ventos (com o Norte para cima)
angles = np.radians([0, 45, 90, 135, 180, 225, 270, 315])

# Criar o gráfico da rosa dos ventos
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plotar os dados de albedo
ax.plot(angles, albedo_means, label='Albedo', linestyle='-', linewidth=2, color = 'seagreen')

# Plotar os dados de razão de Bowen
ax.plot(angles, bowen_ratio_means, label='Razão de Bowen', linestyle='--', linewidth=2, color = 'red')

# Plotar os dados de rugosidade
ax.plot(angles, roughness_means, label='Rugosidade (m)', linewidth=2, color = 'skyblue')

# Adicionar grade radial e definir as legendas
ax.grid(True)
ax.legend()

# Definir os rótulos dos ângulos
ax.set_xticks(angles)
ax.set_xticklabels(octant_labels)

# Inverter a direção do ângulo para que o Norte fique para cima
ax.set_theta_zero_location('N')

# Adicionar título ao gráfico
ax.set_title('Médias de Parâmetros Meteorológicos por Direção do Vento')

# Mostrar o gráfico
plt.show()

# Iterar sobre os octantes de vento
for i, (label, albedo_mean, bowen_ratio_mean, roughness_mean, num_sample) in enumerate(zip(octant_labels, albedo_means, bowen_ratio_means, roughness_means, num_samples)):
    print(f'Octante {label}:')
    print(f'Média de Albedo: {albedo_mean:.2f}')
    print(f'Média de Razão de Bowen: {bowen_ratio_mean:.2f}')
    print(f'Média de Rugosidade (m): {roughness_mean:.2f}')
    print(f'Número de amostras: {num_sample}')
    print()
```

```python
from datetime import datetime

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from netCDF4 import num2date
import numpy as np
import cartopy.io.shapereader

from siphon.catalog import TDSCatalog

# Acessando o catálogo GFS
best_gfs = TDSCatalog('http://thredds.ucar.edu/thredds/catalog/grib/NCEP/GFS/'
                      'Global_0p25deg/catalog.xml?dataset=grib/NCEP/GFS/Global_0p25deg/Best')
print(list(best_gfs.datasets))

# Selecionando o melhor conjunto de dados
best_ds = best_gfs.datasets[0]
ncss = best_ds.subset()

# Criando uma consulta
query = ncss.query()

# Configurando a caixa geográfica para Vidal Ramos
query.lonlat_box(north=-25.8, south=-29.8, east=-48, west=-52).time(datetime.utcnow())
query.accept('netcdf4')
query.variables('Temperature_surface')

# Obtendo os dados
data = ncss.get_data(query)
print(list(data.variables))

# Extraindo a variável de temperatura
temp_var = data.variables['Temperature_surface']

# Extraindo o nome da variável de tempo do atributo de coordenadas da variável de temperatura
time_name = temp_var.coordinates.split()[1]
time_var = data.variables[time_name]
lat_var = data.variables['latitude']
lon_var = data.variables['longitude']

# Extraindo os valores de dados reais e removendo quaisquer dimensões de tamanho 1
temp_vals = temp_var[:].squeeze()
lat_vals = lat_var[:].squeeze()
lon_vals = lon_var[:].squeeze()

# Convertendo o número de horas desde o tempo de referência para uma data real
time_val = num2date(time_var[:].squeeze(), time_var.units)

# Convertendo temperaturas de Kelvin para Celsius
temp_vals = temp_vals - 273.15

# Combinando latitude e longitude 1D em uma grade 2D de locais
lon_2d, lat_2d = np.meshgrid(lon_vals, lat_vals)

# Encontrando o índice mais próximo da coordenada de Vidal Ramos
vidal_ramos_lon, vidal_ramos_lat = -49.3611, -27.3881

lon_idx = (np.abs(lon_vals - vidal_ramos_lon)).argmin()
lat_idx = (np.abs(lat_vals - vidal_ramos_lat)).argmin()

# Extraindo a temperatura na coordenada de Vidal Ramos
vidal_ramos_temp = temp_vals[lat_idx, lon_idx]

# Criando uma nova figura com tamanho reduzido
fig = plt.figure(figsize=(10, 8))

# Adicionando o mapa e configurando a extensão para as novas coordenadas
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent([-52, -48, -29.8, -25.8])

# Carregar os limites dos municípios de Santa Catarina a partir de um arquivo shapefile
municipios_sc = cartopy.io.shapereader.Reader('C:\\Users\\HP\\Documents\\Gabriel.py\\ENS5173\\dados brutos\\MAPA\\SC_UF_2022.shp')

# Adicionar os limites dos municípios ao plot
ax.add_geometries(municipios_sc.geometries(), ccrs.PlateCarree(), edgecolor='black', facecolor='none', linewidth=0.5)

# Contornando a temperatura em cada lat/long
cf = ax.contourf(lon_2d, lat_2d, temp_vals, 200, transform=ccrs.PlateCarree(), zorder=0, cmap='coolwarm')

# Adicionando uma barra de cores para mostrar a temperatura e reduzindo o tamanho dela
plt.colorbar(cf, ax=ax, fraction=0.032, label='Temperatura (°C)')

# Adicionando um título com o valor de tempo
ax.set_title(u'Previsão de temperatura (°C) para {0:%d %B %Y %H:%MZ}'.format(time_val), fontsize=20)

# Plotando marcadores para cada lat/long para mostrar os pontos da grade GFS de 0,25 graus
ax.plot(lon_2d.flatten(), lat_2d.flatten(), marker='o', color='black', markersize=2, alpha=0.3, transform=ccrs.Geodetic(), zorder=2, linestyle='none')

# Adicionando um ponto para a localização de Vidal Ramos
ax.plot(vidal_ramos_lon, vidal_ramos_lat, marker='h', color='darkslategray', markersize=8, transform=ccrs.Geodetic(), zorder=3)
ax.text(vidal_ramos_lon + 0.05, vidal_ramos_lat, f'Vidal Ramos\n{vidal_ramos_temp:.2f} °C', transform=ccrs.Geodetic(), fontsize=11, color='black', weight='bold')

plt.show()
```
