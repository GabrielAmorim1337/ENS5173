```python
#Criando uma lista em python
list = [0,'eu amo polatm', 2.5]
print(list)

#neta parte foi realizada a adiçao de duas strings, sendo que list['] é o segundo indice do vetor list
print('variável:'+list[1])

#adição do primeiro indice da variavel list (list[0]) - lembrar que a indexação em python começa em 0
print(list[0]+1)
      
```

```python
#criando matriz numpy
import numpy as np
#criando uma matriz de zeros
a = np.zeros((5,5))
print(a)
print('Indexação numpy')
print(a[0,0])
```

```python
#Criando números aleatórios e mostrando a dimensão da matriz
b = np.random.rand(5,5)
print(b)
print('Dimensão da matriz')
print(b.shape)
```

```python
#Criando matriz de valores aleatórios entre uma faixa estabelescida
vmin = 1 #valor mínimo
vmax = 100 # valor máximo

c = vmin+(vmax-vmin)*b
print(c)
```

```python
#Plotando matriz
import matplotlib.pyplot as plt

plt.plot(c.transpose())
```

```python
#Plotando apenas a primeira linha
x = np.array(range(0,np.size(c[0,:])))
print(x)
plt.xticks(x)
plt.plot(x,c[0,:])
```

```python
#plotando apenas a primeira coluna
plt.plot(c[:,0])
```

```python
#Criando um gráfico de barras e exportando

fig, ax = plt.subplots(3,1)
ax[0].bar(x,c[:,0])
ax[1].bar(x,c[:,0])
ax[2].bar(x,c[:,0])

# exportar a figura
import os
rootfolder = os.getcwd()
print(rootfolder)
figfolder=rootfolder+r'/figures/'
print(figfolder)
fig.savefig(figfolder+'Figure01.png')
```
